- 결측치 처리 

```python
import seaborn as sns
titanic = sns.load_dataset('titanic')
titanic.head()

# 타이타닉 탑승자 데이터 앞 5개 
```

```python
titanic.info()

# 891이 대부분인데 그보다 작은 데이터는 나머지가 결측치 (age.deck..)
```

```python
titanic.isna().sum()
```

```python
# age 결측치는 age 열의 평균으로 대체 
titanic['age'].fillna(titanic.age.mean(), inplace=True)
titanic.tail(3)

# 비교하기 
```

```python
# embarked 결측치는 행 삭제 (subset=)
titanic.dropna(subset=['embarked'], inplace=True)
titanic.dropna(subset=['embark_town'],inplace=True)
titanic.tail()
```

```python
# deck 컬럼은 컬럼 삭제 (columns=)
# 결측치가 너무 많아서 deck 자체를 날리므로 drop 
titanic.drop(columns=['deck'],inplace=True)
```

- apply ( list의 map)

```python
df
```

```python
# 각 열의 최댓값과 최소값의 차이 
df.apply(lambda x : x.max() - x.min()) # axis=0 생략 

# 데프 전체 / 특정 둘다 적용 
# (함수)
# 값 다른데..암튼 이렇게 쓴다..
```

```python
# 각 행의 최댓값-최솟값 차이 
df.apply(lambda x : x.max() - x.min(), axis=1)
```

```python
# 타이타닉 승객 중 20세 넘으면 'adult' , 그렇지 않으면 'child'
titanic['adult/child'] = titanic.age.apply(lambda x : 'adult' if x.age > 20 else 'child')  # 람다는 한 문장만 작성 /if 작성하고 싶으면 삼항연산자 
titanic.tail()
```

```python
# cctv 소숫점 없애기 

cctv.head()
```

```python
# 총계 연습
cctv['총계'] = cctv['총계'].apply(lambda x : int(x.replace(',','')))       # ,를 먼저 떼주고 정수변환(int) 순서** 
cctv.head()
```

```python
# 그럼 원하는 범위를 하고 싶을 때는(2012년 이전~ 끝)
```

```python
cctv.columns        
# 리스트 형태로 출력됨 
# 바꾸고 싶은 건 2012년 이전(인덱스 값 2~ ) - 
# 리스트니까 슬라이스 
```

```python
for column in cctv.columns[2:]:
    cctv[column] = cctv[column].apply(lambda x : int(x.replace(',','')))
cctv.head()
```
