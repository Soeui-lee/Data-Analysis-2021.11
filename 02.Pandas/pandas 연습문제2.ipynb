- b. 3분위수(Q3)와 1분위수(Q1)의 차이보다 1.5배가 크거나 작은 데이터는 이상치이다.
- 즉, Q1 - 1.5 * (Q3 - Q1) 보다 작은 데이터
- 이 이상치를 제거하고 위의 4가지 항목에 대해서 평균, 표준편차를 구하시오.

```python
# 이상치 
s= iris[iris.species == 'setosa']['sepal_width']
q1, q3 =s.quantile(0.25), s.quantile(0.75)
iqr = q3 - q1
q1, q3, iqr
```

```python
# 상한치/하한치 구하기 
upper = q3 +1.5 * iqr
lower = q1 -1.5 * iqr
# upper 보다 크거나 lower 보다 작은 s값은 이상치  
```

```python
# '이상치 제거' : 값만 제거할 수 X (행, 열 삭제가 아니니까 )
# NaN로 만들어주면 제거하고 계산 가능  
```

```python
s[(s < lower) | (s > upper)] = np.nan # | : 버티컬...라인? 
s.isna().sum()
```

```python
s.mean(), s.std()
```

-위 같은 과정을 계산해 s.isna( ).sum( ), s.mean( ), s.std() 를 반환하는 함수 만들기 
(다시 )

```python
def get_new_stat(s):
    q1, q3 =s.quantile(0.25), s.quantile(0.75)
    iqr = q3 - q1
    upper = q3 +1.5 * iqr
    lower = q1 -1.5 * iqr
    s[(s < lower) | (s > upper)] = np.nan # | : 버티컬...라인? 
    return s.isna().sum, s.mean, s.std()
  
```

### 2. Titanic

- a. 타이타닉호의 승객에 대해 나이와 성별에 의한 카테고리 열인 category1 열을 만드시오.
- category1 카테고리는 다음과 같이 정의됨
- 1) 20살이 넘으면 성별을 그대로 사용한다.
- 2) 20살 미만이면 성별에 관계없이 “child”라고 한다.

```python
# 문제를 보면 or 임 (3항연산자)
# T 데프에 apply 했으므로 X 는 컬럼
# titanic. 했으면 이 x는 엘리먼트(?)
# axis = 1 : 해당 방향으로 같은 사람의 데이터로만 계산해야하니까(조작에서의 나이child/adult는 age만,그리고 element뿐이었어서 X)
titanic['category1'] = titanic.apply(lambda x : x.sex if x.age>= 20 else 'child', axis = 1)
titanic.tail()
```

- 타이타닉호의 승객 중 나이를 명시하지 않은 고객은 나이를 명시한 고객의 평균 나이
값이 되도록 titanic 데이터프레임을 고치시오.

```python
titanic['age'] = titanic.age.fillna(titanic.age.mean())
titanic['category1'] = titanic.apply(lambda x : x.sex if x.age >=20 else 'child', axis = 1)
titanic.tail()
```

- 성별, 선실(class)별, 출발지(embark_town)별 생존율을 구하시오.

```python
# survived의 평균 ---- 생존률 
titanic.groupby('sex')['survived'].mean()
```

- a. 배기량(displacement) 대비 마력(horsepower) 열(hp_per_cc)을 추가하시오.

```python
mpg['hp_per_cc'] = mpg.apply(lambda x : x.displacement/x.horsepower, axis =1)
mpg.head()
```

```python
mpg['manufacturer'] = mpg.name.apply(lambda x : x.split()[0])
mpg['model'] = mpg.name.apply(lambda x : ' '.join(x.split()[1:]))
mpg.drop(column=['name'], inplace = True)
mpg.head()
```
