# Matplotlib - 여러개의 그래프

```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rcParams['axes.unicode_minus'] = False 
plt.rcParams['font.family'] = 'Malgun Gothic'
```

```python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)
def g(t):
    return np.sin(np.pi*t)
t1 = np.arange(0.0, 5.0, 0.01)
t2 = np.arange(0.0, 5.0, 0.01)          

plt.subplot(221)                # 행2 x 열2이고, 그래프 중 첫번째 ( 221 or 2,2,1 표현 )
plt.plot(t1, f(t1))
plt.subplot(2, 2, 2)
plt.plot(t2, g(t2))
plt.subplot(223)
plt.plot(t1, f(t1), 'r-')
plt.subplot(224)
plt.plot(t2, g(t2), 'r-')
plt.show()
```

```python
import seaborn as sns
ans = sns.load_dataset('anscombe')
ans.head(3)
```

```python
ds = ans[ans.dataset == 'I']
plt.scatter(ds.x, ds.y, marker = 'o')
plt.title('Dataset I')
plt.show()
```

```python
# 첫 ~ 끝 점 x는 4 와 14 
# 식에 대입해보면 y = 5와 10 
```

```python
# 하나를 선택해서 연습게임
ds = ans[ans.dataset == 'I']
plt.scatter(ds.x, ds.y, marker = 'o')
plt.plot([4,14],[5,10], 'r-', lw =2)
plt.title('Dataset I')
plt.show()
```

```python
fig, axes = plt.subplots(2,2,figsize=(10,8))
for index, roman in enumerate(ans.dataset.unique()):    #['I','II','III','IV'] 로만... 
    ds = ans[ans.dataset == roman]
    ax = axes[index//2, index%2]
    ax.scatter(ds.x, ds.y, marker = 'o')
    ax.plot([4,14],[5,10], 'r-', lw =2)
    ax.set_title(f'Dataset {roman}')
plt.show()

```

```python
# 조정해주기 
fig, axes = plt.subplots(2,2,figsize=(10,8))
for index, roman in enumerate(ans.dataset.unique()):    #['I','II','III','IV'] 로만... 
    ds = ans[ans.dataset == roman]
    ax = axes[index//2, index%2]
    ax.scatter(ds.x, ds.y, marker = 'o')
    ax.plot([3,20],[4.5,13], 'r-', lw =2)
    ax.set_title(f'Dataset {roman}')
plt.show()
```

```python
# 제목달아주기 
fig, axes = plt.subplots(2,2,figsize=(10,8))
for index, roman in enumerate(ans.dataset.unique()):    #['I','II','III','IV'] 로만... 
    ds = ans[ans.dataset == roman]
    ax = axes[index//2, index%2]
    ax.scatter(ds.x, ds.y, marker = 'o')
    ax.plot([3,20],[4.5,13], 'r-', lw =2)
    ax.set_title(f'Dataset {roman}')
fig.suptitle("Anscombe's Quartet", fontsize=16 )
plt.show()
```
