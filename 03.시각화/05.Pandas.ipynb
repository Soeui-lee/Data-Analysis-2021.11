# Pandas 시각화

```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rcParams['axes.unicode_minus'] = False 
plt.rcParams['font.family'] = 'Malgun Gothic'
```

- 특정일~ n일동안의 데이터 시각화

```python
np.random.seed(2021)
df = pd.DataFrame(np.random.randn(100,3),       # 100개 꺼내되 3줄로
                    index=pd.date_range('1/1/2021', periods = 100),     #날짜라는 걸 알게 함 
                    columns = list('ABC')).cumsum()  # cumulate+sum 계속해서 더해감
df.plot()
plt.title('Pandas의 plot method 사용 예')
plt.xlabel('일자'), plt.ylabel('데이터')
plt.show()
```

### 2. Bar chart

```python
import seaborn as sns
iris = sns.load_dataset('iris')

plt.figure(figsize=(12,6))
iris.sepal_length[:50].plot(kind='bar',rot=0) #원하는 plot 형태 지정 (kind =) / # rot =0 : 인덱스(x축 숫자들) 바로 세우기(means 90도) 
plt.title('Setosa 품종의 Sepal length')
plt.show()
```

```python
df2 = iris.groupby('species').mean()
df2.columns.name = 'Feature'        # 나중에 레전드에 사용 
df2
```

```python
df2.plot.bar(rot=0)     # kind = 'bar'의 다른 형태 
plt.title('Iris 각 품종의 Feature별 평균')
plt.show()
```

```python
# 사이즈 늘리기
df2.plot.bar(rot=0)    
plt.title('Iris 각 품종의 Feature별 평균')
plt.ylim(0,9)       # y축의 길이를 늘리기, y limit를 정해주기
plt.show()
```

### 3. Pie chart

```python
titanic = sns.load_dataset('titanic')
s = titanic['class'].value_counts()
s
```

```python
s.plot.pie(autopct='%.2f%%')    # auto percent( 양 끝% :포매팅),(중간 % : pct),(.2f : 소수점 둘째자리까지 표시)
plt.title('ahk')
plt.show()
```

### 4. Box plot

```python
iris[iris.species == 'setosa'].plot.box()
plt.title('Setosa 품종의 각 Feature별 Box plot')
plt.show()
```

```python
# 사이즈 바꾸기
iris[iris.species == 'setosa'].plot.box(figsize=(8,10))
plt.title('Setosa 품종의 각 Feature별 Box plot')
plt.show()

# 그럼 다양한 플롯에 있는 그래프랑 똑같음 (확인해봐! )
```

```python
iris.boxplot(by='species', figsize=(10,8))
plt.title('각 피쳐의 품종별 데이터에 대한 box plot')
plt.show()
```
