# Matplotlib - 다양한 plot

```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rcParams['axes.unicode_minus'] = False 
plt.rcParams['font.family'] = 'Malgun Gothic'
```

### 1. 산점도 (Scatter plot)

```python
import seaborn as sns
ans = sns.load_dataset('anscombe')
ans.head(3)
```

```python
ans.groupby('dataset').agg(['mean','std'])

# 얼핏 보면 4개의 데이터 전부 같다고 착각할 수 
```

```python
import random
```

```python
X,Y = np.random.randn(100), np.random.randn(100)
plt.title('산점도')
plt.scatter(X,Y)
plt.show()
```

```python
colors = np.random.rand(100)       # rand : 0~1
area = np.pi * (15 * np.random.rand(100))**2
plt.scatter(X,Y, s= area, c=colors, alpha=0.5) # 사이즈. 칼라, 투명도 
plt.title('Bubble chart')
plt.show()
```

### 2. Bar Chart ( 막대 그래프 )

- x축이 카테고리값일 때 사용 

```python
s = pd.Series(np.random.rand(10), index=list('abcdefghij'))
plt.bar(s.index,s.values)           # x축이 인덱스, y축이 values 
plt.title('Bar chart')
plt.show()
```

- 타이타닉 활용

```python
import seaborn as sns 
titanic = sns.load_dataset('titanic')
titanic['class'].value_counts()     # class가 키워드라서 titanic.class 불가하다고 
# 값이 시리즈로 나오니까 활용하면 됨 
```

```python
s = titanic['class'].value_counts() 
plt.bar(s.index,s.values)          
plt.title('타이타닉호의 선실별 승객수')
plt.show()
```

```python
# bar chart 옆으로 나타내기 (barh)
s = pd.Series(np.random.rand(10), index=list('abcdefghij'))
plt.barh(s.index,s.values)          
plt.title('Bar chart')
plt.show()
```

### 3. Histogram

- 데이터의 분포를 한눈에

```python
s = np.random.randn(200)
plt.hist(s)
plt.title('Histogram')
plt.show()

# bins 기본값 = 10 
```

```python
plt.hist(s, bins=20)
plt.title('bins=20인 경우')
plt.show()
```

### 4. Box plot 

- 통계량 확인 시 

```python
# Setosa 품종
iris = sns.load_dataset('iris')
setosa = iris[iris.species == 'setosa']
```

```python
plt.figure(figsize=(8,10))
plt.boxplot([setosa.sepal_length, setosa.sepal_width,
            setosa.petal_length,setosa.petal.width],
            label= iris.columns[:-1])
plt.title('Setosa 품종의 각 피쳐별 박스 플롯', fontsize=15)
plt.show()

# 왜 오류가 뜨지?
```

```python
plt.figure(figsize=(8,10))
plt.boxplot([setosa.sepal_length, setosa.sepal_width,
setosa.petal_length, setosa.petal_width],
labels=['sl','sw','pl','pw'])

plt.title('Setosa 품종의 각 피쳐', fontsize=15)
plt.show()

# 범위를 벗어난 이상치는 동그라미로 표시
# 책 코드 

```

### 5. 파이(Pie) 차트

- 범주별 구성 비율 

```python
ratio = [36,32,16,18] # 백분률
labels = ['사과','바나나','멜론','포도']

plt.pie(ratio, labels = labels, autopct='%.1f%%')   # 퍼센트는 소숫점 첫번째까지 보여달라 
plt.show()

# 그래프는 반시계 방향으로 읽음 
```
