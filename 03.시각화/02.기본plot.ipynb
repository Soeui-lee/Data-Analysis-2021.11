# Matplotlib - plot 기본

```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rcParams['axes.unicode_minus'] = False 
plt.rcParams['font.family'] = 'Malgun Gothic'
```

```python
plt.plot([0,1,2,3], [0,1,4,9])   
plt.title('라인플롯')           
plt.grid()                      
plt.show()                      
```

- Line  style

```python
# 옵션 주기 
plt.plot([0,1,2,3], [0,1,4,9], 'ro-')   # 'r'ed , circle(o), 직선line(-) : color-marker-line style 순 
plt.title('red-circle-line')           
plt.grid()                      
plt.show()  
```

```python
plt.plot([0,1,2,3], [0,1,4,9], 'ms--')   
plt.title('Magenta-Square-Dashed')           
plt.grid()                      
plt.show()  
```

```python
plt.plot([0,1,2,3], [0,1,4,9], 'ms--')   
plt.plot([8,5,2,1], 'g^:')      #리스트가 하나만 주어지면 y 로 인식 
plt.title('Magenta-Square-Dashed, Green-삼각형-점선')           
plt.grid()                      
plt.show()  
```

```python
plt.plot([0,1,2,3], [0,1,4,9], 'ms--',linewidth=3,markersize=10)   
plt.plot([8,5,2,1], 'g^:', lw=5, ms = 20)     
plt.title('Line Style')           
plt.grid()                      
plt.show() 
```

- 그림 범위 지정 

```python

plt.plot([0,1,2,3], [0,1,4,9], 'ro-') # color, marker, linestyle 
plt.title('X축, Y축 범위 지정')   
plt.xlim(-1,5)          # x limit 
plt.ylim(-1,10)         # y limit 
plt.grid()                      
plt.show()  
```

- Tick 설정 

```python
X = np.linspace(-np.pi, np.pi, 500)
Y = np.cos(X)
plt.title('Tick 설정')
plt.plot(X,Y)
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi])
plt.yticks([-1,0,1])
plt.grid()
plt.show()
```

- 범례(Legend) # 안내

```python
X = np.linspace(-np.pi, np.pi, 500)
C, S = np.cos(X), np.sin(X)
plt.title('Legend를 표시한 플롯')
plt.plot(X,C,label='cosine')
plt.plot(X,S,label='sine')
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi])
plt.yticks([-1,0,1])
plt.grid()
plt.legend()        # legend와 label만 있으면 O 
plt.show()
```

- X축, Y축 Label 

```python
X = np.linspace(-np.pi, np.pi, 500)
C, S = np.cos(X), np.sin(X)
plt.title('X축, Y축 라벨을 표시한 플롯')
plt.plot(X,C,label='cosine')
plt.plot(X,S,label='sine')
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi])
plt.yticks([-1,0,1])
plt.grid()
plt.legend()        
plt.xlabel('time'), plt.ylabel('Amplitude')         #',' 사용해서 작성 가능 
plt.show()
```

- 그림의 크기

```python
plt.figure(figsize=(10,6))
plt.title('그림크기 지정')
plt.plot(X,C,label='cosine')
plt.plot(X,S,label='sine')
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi])
plt.yticks([-1,0,1])
plt.grid()
plt.legend()        
plt.xlabel('time'), plt.ylabel('Amplitude')
plt.show()
```

- 그림 저장

```python
plt.figure(figsize=(10,6))
plt.title('그림크기 지정한 그래프를 저장')
plt.plot(X,C,label='cosine')        # x,y 자리는 인덱싱으로 꺼내올 수 있는 형태면 가능
plt.plot(X,S,label='sine')
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi])
plt.yticks([-1,0,1])
plt.grid()
plt.legend()        
plt.xlabel('time'), plt.ylabel('Amplitude')
plt.savefig('삼각함수.png')
```
